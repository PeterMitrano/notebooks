```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
```

```python
W = 100
start = [5,5]
goal = [95,95]
step_size = 1
obstacles = [[[10,10], [10,10]], [[20,50], [50,20]], [[80,30], [5,20]]]
```

```python
class Node:
    
    def __init__(self, pos):
        self.pos = pos
        self.children = []
        self.parent = None
    
    def add(self, n):
        assert isinstance(n, Node)
        n.parent = self
        self.children.append(n) 
        
    def __getitem__(self, index):
        return self.pos[index]
    
    def __iter__(self):
        for child in self.children:
            for item in child:
                yield item
        yield self
        
    def __repr__(self):
        return "[{:d},{:d}]".format(self.pos[0], self.pos[1]) + " ".join([i.__repr__() for i in self.children])
```

```python
tree = Node(start)
n = tree
for iterations in range(1000):
    p = list(np.random.randint(0, W, size=2))
    
    for i in tree:
        if i.pos == p:
            continue
    
    # find closest node to p in tree
    min_d = np.sqrt(2)*W
    for n in tree:
        d = np.sqrt((n[0]-p[0])**2 + (n[1]-p[1])**2)
        if d < min_d:
            min_d = d
            min_n = n
    # check if min_n can be added to the tree
    intersects = False
    for o in obstacles:
        # this is just a lazy hack
        nx = p[0]
        ny = p[1]
        ox1 = o[0][0]
        oy1 = o[0][1]
        ox2 = o[0][0] + o[1][0]
        oy2 = o[0][1] + o[1][1]
        if ox1 < nx < ox2 and oy1 < ny < oy2:
            intersects = True
    if not intersects:
        p_node = Node(p)
        min_n.add(p_node)
       
    d_to_goal = np.sqrt((p[0]-goal[0])**2 + (p[1]-goal[1])**2)
    if d_to_goal <= 2:
        print("Goal reached in {:d} iterations".format(iterations))
        break
        
# Follow path back from goal to start
f = p_node
path = [goal]
while f.parent:
    path.append(f.pos)
    f = f.parent
path.append(start)
path = np.array(path)
```

```python
plt.figure(figsize=(8,8))
ax = plt.gca()
for o in obstacles:
    rect = plt.Rectangle(o[0], width=o[1][0], height=o[1][1], fc='y')
    ax.add_patch(rect)

for i in tree:
    ax.add_patch(plt.Circle(i.pos, radius=1, fc='gray'))
    if i.parent:
        plt.plot([i.pos[0], i.parent.pos[0]], [i.pos[1], i.parent.pos[1]], c='gray', linewidth=6)

plt.plot(path[:,0], path[:,1], c='r',linewidth=1)

ax.add_patch(plt.Circle(start, radius=1, fc='k'))
ax.add_patch(plt.Circle(goal, radius=1, fc='g'))

plt.xlim(0, W)
plt.ylim(0, W)
_ = plt.axis('scaled')
```
